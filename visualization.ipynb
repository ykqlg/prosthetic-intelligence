```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
df = pd.read_csv('data.csv')

# 
ACC_X = df['ACC_X']
ACC_Y = df['ACC_Y']
ACC_Z = df['ACC_Z']

# 
plt.figure(figsize=(8, 4))
plt.plot(ACC_X)
plt.title('ACC_X')

# 
plt.figure(figsize=(8, 4))
plt.plot(ACC_Y)
plt.title('ACC_Y')

# 
plt.figure(figsize=(8, 4))
plt.plot(ACC_Z)
plt.title('ACC_Z')

# 
plt.show()
```
