```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
df = pd.read_csv('data.csv')

# 提取列数据
ACC_X = df['ACC_X']
ACC_Y = df['ACC_Y']
ACC_Z = df['ACC_Z']

# 绘制第一张图
plt.figure(figsize=(8, 4))
plt.plot(ACC_X)
plt.title('ACC_X')
plt.ylim(-3000, 3000)
plt.xlabel('ms')
plt.ylabel('mg')

# 绘制第二张图
plt.figure(figsize=(8, 4))
plt.plot(ACC_Y)
plt.title('ACC_Y')
plt.xlabel('ms')
plt.ylabel('mg')
plt.ylim(-3000, 3000)

# 绘制第三张图
plt.figure(figsize=(8, 4))
plt.plot(ACC_Z)
plt.title('ACC_Z')
plt.xlabel('ms')
plt.ylabel('mg')
plt.ylim(-3000, 3000)

# 显示图形
plt.show()
```
