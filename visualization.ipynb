```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
def get_latest_file(folder_path):
    # 使用 glob 模块获取文件夹中所有文件的路径列表
    files = glob.glob(os.path.join(folder_path, '*'))

    # 如果文件列表为空，返回 None
    if not files:
        return None

    # 获取文件的最后修改时间并排序
    latest_file = max(files, key=os.path.getctime)

    return latest_file


folder_path = './output'
latest_file = get_latest_file(folder_path)
if latest_file:
    print(f"The latest file in {folder_path} is: {latest_file}")
else:
    print(f"No files found in {folder_path}")
```

```python
folder_path = './output'
latest_file = get_latest_file(folder_path)
latest_file = './output/20231226_090955.csv'
df = pd.read_csv(latest_file)

# 提取列数据
ACC_X = df['ACC_X']
ACC_Y = df['ACC_Y']
ACC_Z = df['ACC_Z']
# 创建一个包含三个子图的画布
fig, axes = plt.subplots(1, 3, figsize=(20, 4)) 
irange = 5000
ylim = False
# 绘制第一张图
axes[0].plot(ACC_X)
axes[0].set_title('ACC_X')
axes[1].plot(ACC_Y)
axes[1].set_title('ACC_Y')
axes[2].plot(ACC_Z)
axes[2].set_title('ACC_Z')
for axe in axes:
    if ylim :axe.set_ylim(-irange, irange)
    # axe.set_xlabel('ms')
    axe.set_ylabel('mg')


plt.tight_layout()
plt.show()
```
