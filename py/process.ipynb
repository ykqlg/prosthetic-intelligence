```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./output/1711417679.599828.csv')
data = df['ACC_Z'].values
time = df['Time'].values

differences = np.diff(time)
print(f"sample_rate:{1/np.mean(differences)}")

# 绘制直方图
plt.hist(1/differences, bins=10, color='skyblue', edgecolor='black')

# 添加标题和标签
plt.title('Histogram of Timestamp Differences')
plt.xlabel('Difference')
plt.ylabel('Frequency')

# 显示图形
plt.show()
```
