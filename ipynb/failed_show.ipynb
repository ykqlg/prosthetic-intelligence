```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

label_df = pd.read_csv("./label_file.csv")
label_df = label_df[label_df['Label'] == 0]
file_list = label_df['FilePath'].values
print(len(file_list))
for i, file in enumerate(file_list):
    df = pd.read_csv(file)
    data = df['ACC_Z']
    time = df['Time']
    if i == 3: break
    plt.plot(time,data/1000)
    plt.show()
```
